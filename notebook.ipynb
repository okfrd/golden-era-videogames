The database contains two tables. Each table has been limited to 400 rows for this project, but you can find the complete dataset with over 13,000 games on Kaggle.

### `game_sales` table

| Column | Definition | Data Type |
|-|-|-|  
|name|Name of the video game|`varchar`|

|publisher|Game publisher|`varchar`|

### `reviews` table

| Column | Definition | Data Type |
|-|-|-|
|name|Name of the video game|`varchar`|

|user_score|User score according to Metacritic|`float`|


### `users_avg_year_rating` table

| Column | Definition | Data Type |
|-|-|-|
|year| Release year of the games reviewed |`int`|  
|num_games| Number of games released that year |`int`|
|avg_user_score| Average score of all the games ratings for the year |`float`|

### `critics_avg_year_rating` table

| Column | Definition | Data Type |
|-|-|-|
|year| Release year of the games reviewed |`int`|  
|num_games| Number of games released that year |`int`|
|avg_critic_score| Average score of all the games ratings for the year |`float`|



## 1. Best-Selling Games
Description: This query identifies the top 10 best-selling games by sorting the game_sales table in descending order based on the games_sold column. It ensures that we capture the most commercially successful games in the dataset.

Key Points:
- Used ORDER BY games_sold DESC to prioritize games with the highest sales.
- Limited the result to 10 rows with LIMIT 10.

```python
-- best_selling_games
SELECT *
FROM game_sales
ORDER BY games_sold DESC
LIMIT 10;
```

## 2. Top Years by Critic Scores
The query finds the 10 years with the highest average critic scores, including only years where at least 4 games were released. The average critic score is rounded to 2 decimal places for clarity.

Key Points:

- Included a filter with WHERE num_games >= 4 to ensure a significant sample size.
- Used ROUND(avg_critic_score, 2) for better presentation.
- Results are ordered by the critic score in descending order to highlight the top years.

```python
-- critics_top_ten_years
SELECT year,
		num_games,
		ROUND(avg_critic_score, 2) AS avg_critic_score
FROM critics_avg_year_rating
WHERE num_games >= 4
ORDER BY avg_critic_score DESC
LIMIT 10;
```

## Query 3: Golden Years
This query identifies years where critics and users highly rated the games released. It combines data from users_avg_year_rating and critics_avg_year_rating, filtering for years where either the average user score or critic score was greater than 9.

Key Points:
- Performed a JOIN between the users_avg_year_rating and critics_avg_year_rating tables based on the year column.
- Added a calculated column, diff, showing the difference between the average critic and user scores.
- Results are ordered by year in ascending order for chronological analysis.


```python
-- golden_years
SELECT users_avg_year_rating.year, users_avg_year_rating.num_games, users_avg_year_rating.avg_user_score, critics_avg_year_rating.avg_critic_score, (critics_avg_year_rating.avg_critic_score - users_avg_year_rating.avg_user_score) AS diff
FROM users_avg_year_rating
JOIN critics_avg_year_rating
ON users_avg_year_rating.year = critics_avg_year_rating.year
WHERE users_avg_year_rating.avg_user_score > 9
OR critics_avg_year_rating.avg_critic_score > 9
ORDER BY users_avg_year_rating.year ASC;
```
